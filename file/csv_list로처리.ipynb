```python
import csv

# 파일 오픈
file = open('./card.csv')
# 파일 읽기 reader
data = csv.reader(file)
# 방법 1) data의 첫번째 행이 제목이여서 해당부분 스킵해서 활용시. 
# next(data)
# data = list(data)

# 방법 2) 제목과 데이터 변수로 담을시
fileData = list(data)
title = fileData[0]
data = fileData[1:]
data
```

```python
이용금액=[]
# for문 활용
# for row in data:
#     이용금액.append(int(row[6]))

# list comprehension
이용금액 = [int(row[6]) for row in data]

## 값
print('전체 합 {}, 최소 {}, 최대 {}, 평균 {}'.format(
    sum(이용금액),
    min(이용금액),
    max(이용금액),
    sum(이용금액) / len(이용금액)
))
```

```python
#####################################
## 산정도를 통해서 데이터의 분포를 확인.
##################################### 

import matplotlib.pyplot as plt
plt.plot(이용금액, 'o')  
```

```python
# 점표매입에 관련된 이용금액만

plt.rcParams['font.family'] = 'Malgun Gothic' # 폰트가 있어야지 안 깨짐

전표매입리스트 = []
for row in data:
    if row[-1] == '전표매입':
        전표매입리스트.append(int(row[6]))

전표취소리스트 = []
for row in data:
    if row[-1] == '전표취소':
        전표취소리스트.append(int(row[6]))
        
승인취소리스트 = []
for row in data:
    if row[-1] == '매출취소':
        전표취소리스트.append(int(row[6]))

전표구분딕셔너리 = {
    '전표매입리스트': 전표매입리스트,
    '매출취소이용금액': 전표취소리스트,
    '승인취소이용금액': 승인취소리스트
}

import matplotlib.pyplot as plt
plt.plot(전표구분딕셔너리['전표매입리스트'], 'o')
plt.title('전표매입한 이용금액')
```
